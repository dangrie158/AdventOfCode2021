```python
import numpy as np
from itertools import product
```

```python
with open("inputs/11.txt") as infile:
    initial_energy_map = np.array([[int(x) for x in line]for line in infile.read().splitlines()], dtype=np.int8)
```

```python
def make_step(energy_map):
    energy_map += 1
    lights_flashed = np.full(energy_map.shape, False)
    while np.any((energy_map > 9) & np.logical_not(lights_flashed)):
        flashing_lights = np.where((energy_map > 9) & np.logical_not(lights_flashed))
        for x, y in zip(*flashing_lights):
            lights_flashed[x, y] = True

            x_indices = range(max(0, x-1), min(x+2, energy_map.shape[0]))
            y_indices = range(max(0, y-1), min(y+2, energy_map.shape[1]))
            indices = list(product(x_indices, y_indices))
            indices.remove((x, y))
            window = tuple(zip(*indices))
            energy_map[window] += 1

    energy_map[np.where(energy_map > 9)] = 0
    return lights_flashed

energy_map = initial_energy_map.copy()
total_flashes = 0
for step in range(100):
    lights_flashed = make_step(energy_map)
    total_flashes += np.sum(lights_flashed)

print(f"{total_flashes=}")
```
